```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Student_scores.csv")
```

```python
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
# to check how many null values is null in dataset
df.isnull().sum()
```

```python
# drop unnamed column
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

# change weekly study hours column

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

# gender distribution

```python
# for figure size means to look very clean height and width
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

# from the above chart
#we have the no of females in data is more than the no of males

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
# use heatmap for reltaionship graphs
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Realationship between Parent's Education and Student Score")
plt.show()
```

```python
# From the above chart 
# we have concluded that the education of the parents have a good impact on students marks
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Realationship between Parent's Marital Status  and Student Score")
plt.show()
```

```python
# from the above chart
# we have Concluded that there is no/negligible impact on ParentMaritalStatus on Student Marks
```

```python
print(df['EthnicGroup'].unique())
```

# Fill the missing values in dataset like this

```python
# use this like 
# df['col1'] = df['col1'].fillna(df['col1'].mode()[0])
df["EthnicGroup"]=df["EthnicGroup"].fillna(df["EthnicGroup"].mode()[0])
print(df['EthnicGroup'].unique())
```

# Distribution of Ethnic Groups

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
print(groupA)

print(groupA["EthnicGroup"])
```

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()

# labels
l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

print(mlist)
# use any aextra clear use explore and legends
plt.pie(mlist,labels=l,autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

```python
ax=sns.countplot(data=df,x="EthnicGroup")
ax.bar_label(ax.containers[0])
plt.title("Distribution on Ethnic Group")
```

```python
# from above chart 
# we have concluded that ethinic group C has very much impact on student marks
```
